```python
X = df.drop("Label", axis = 1)
y = df["Label"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
scaler = StandardScaler()
knn = KNeighborsClassifier()
```

```python
operations = [("scaler",scaler), ("knn",knn)]
```

```python
pipe = Pipeline(operations)
```

```python
k_values = list(range(1, 30))
```

```python
param_grid = {"knn__n_neighbors":k_values}
```

```python
# knn()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = GridSearchCV(pipe, param_grid=param_grid, scoring="accuracy", cv = 5)
```

```python
model.fit(X_train, y_train)
```

```python
model.best_estimator_.get_params()
```

```python
plt.plot(k_values, model.cv_results_["mean_test_score"], "bo", ls= "-")
plt.xlabel("k")
plt.ylabel("Accuracy")
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay, plot_confusion_matrix
```

```python
y_pred = model.predict(X_test)
```
