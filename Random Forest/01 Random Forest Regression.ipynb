## Linear Reg

```python
from sklearn.linear_model import LinearRegression
```

```python
lrmodel = LinearRegression()
```

```python
lrmodel.fit(X_train, y_train)
```

```python
predlr = lrmodel.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,predlr))
```

```python
signal_range = np.arange(0, 100)
```

```python
pred_lr = lrmodel.predict(signal_range.reshape(-1,1))
```

## Polynomial Reg

```python
model = LinearRegression()
```

```python
def run_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    
    pred = model.predict(X_test)
    
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    print(f"RMSE: {rmse}")
    
    signal_range = np.arange(0, 100)
    pred_ = model.predict(signal_range.reshape(-1,1))
    
    plt.figure(figsize=(10 ,7), dpi = 130)
    sns.scatterplot(data = df, x = df.signal, y = df.density)
    plt.plot(signal_range, pred_)
```

```python
run_model(model, X_train, X_test, y_train, y_test)
```

### Pipeline for various degree

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
pipe = make_pipeline(PolynomialFeatures(2),LinearRegression())
```

```python
run_model(pipe,X_train,X_test,y_train, y_test)
```

## KNN Reg

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn  = KNeighborsRegressor(n_neighbors=5)
run_model(knn,X_train,X_test,y_train, y_test)
```

```python
k_val = [6, 10, 19]
for i in k_val:
    knn  = KNeighborsRegressor(i)
    print(f"MODEL FOR KVAL OF {i}")
    run_model(knn,X_train,X_test,y_train, y_test)
    print("\n")
```

## Decision Tree Reg

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(max_depth=10, max_leaf_nodes=130)
run_model(model,X_train,X_test,y_train, y_test)
```

## Support Vector Regression

```python
from sklearn.svm import SVR
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {"C":[0.001, 0.01, 0.1, 0.8, 1, 100, 1000], "gamma":["auto", "scale"]}
svr = SVR()
```

```python
model = GridSearchCV(svr, param_grid)
```

```python
print(model.best_params_)
run_model(model,X_train,X_test,y_train, y_test)
```

```python
model.best_estimator_
```
